```python
import numpy as np
import math
import matplotlib.pyplot as plt
import cv2
from mpl_toolkits.mplot3d import Axes3D
def create_gaussian_kernel(size,sigma):
    kernel=np.zeros((size,size))
    center=size//2
    constant=1/(2*math.pi*sigma**2)
    for x in range(size):
        for y in range(size):
            kernel[x, y] = constant * math.exp(-((x - center) ** 2 + (y - center) ** 2) / (2 * sigma ** 2))
    kernel/=np.sum(kernel)
    return kernel
size=int(input("size=:"))
sigma=float(input("sigma=:"))
gaussian_kernel = create_gaussian_kernel(size, sigma)
plt.imshow(gaussian_kernel, cmap='viridis', interpolation='nearest')
plt.title('Gaussian Kernel')
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(0, size, 1)
y = np.arange(0, size, 1)
x, y = np.meshgrid(x, y)
ax.plot_surface(x, y, gaussian_kernel, cmap='viridis')
plt.title('Gaussian Kernel')
plt.show()
```

```python
gaussian_kernel2 = cv2.getGaussianKernel(size, sigma)
gaussian_kernel2 = np.outer(gaussian_kernel2, gaussian_kernel2)
plt.imshow(gaussian_kernel2, cmap='viridis', interpolation='nearest')
plt.title('Gaussian Kernel2')
plt.show()
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(0, size, 1)
y = np.arange(0, size, 1)
x, y = np.meshgrid(x, y)
ax.plot_surface(x, y, gaussian_kernel2, cmap='viridis')
plt.title('Gaussian Kernel')
plt.show()
```

```python
file_path=input()
image=cv2.imread(f'{file_path}')
def gaussian_blur(image,kernel):
    image2=np.zeros_like(image)
    image = cv2.copyMakeBorder(image, kernel.shape[0] // 2, kernel.shape[0] // 2, kernel.shape[1] // 2, kernel.shape[1] // 2, cv2.BORDER_REFLECT)
    for c in range(image2.shape[2]):
        for x in range(image2.shape[0]):
            for y in range(image2.shape[1]):
                roi = image[x:x + kernel.shape[0], y:y + kernel.shape[1]]
                image2[x, y, c] = (roi[:, :, c] * kernel).sum()
    return image2
image2=gaussian_blur(image,gaussian_kernel2)
cv2.imwrite('blurred_image.jpg', image2)
cv2.imshow('Original Image', image)
cv2.imshow('Blurred Image', image2)
```
